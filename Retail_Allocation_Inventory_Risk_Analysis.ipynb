# Step 02 Inventory Coverage Analysis

### Objective

Evaluate how long current inventory can support forecast demand and identify potential inventory imbalance across products and warehouses.

### 2.1 Calculate Inventory Coverage

```python
df["Inventory_Coverage"] = (
    df["Inventory_Level"]
    /
    df["Demand_Forecast"]
)

df["Inventory_Coverage"].describe()
```

```python
(df["Demand_Forecast"] == 0).sum()
df[df["Demand_Forecast"] == 0]
```

```python
(df["Demand_Forecast"] == 0).sum()
```

### 2.2 Handle Zero Forecast Records

```python
df["Inventory_Coverage_Clean"] = np.where(
    df["Demand_Forecast"] > 0,
    df["Inventory_Level"] / df["Demand_Forecast"],
    np.nan
)
df["Inventory_Coverage_Clean"].describe()
```

### 2.3 Investigate Extreme Coverage Values

```python
df.sort_values(
    by="Inventory_Coverage_Clean",
    ascending=False
)[[
    "SKU_ID",
    "Warehouse_ID",
    "Inventory_Level",
    "Demand_Forecast",
    "Inventory_Coverage_Clean"
]].head(10)
```

### 2.4 Inventory Coverage Distribution

```python
df["Inventory_Coverage_Clean"].quantile(
    [0.25, 0.5, 0.75, 0.9, 0.95, 0.99]
)
```

### 2.5 Inventory Health Classification

```python
conditions = [
    df["Inventory_Coverage_Clean"] < 14,

    (df["Inventory_Coverage_Clean"] >= 14) &
    (df["Inventory_Coverage_Clean"] < 30),

    (df["Inventory_Coverage_Clean"] >= 30) &
    (df["Inventory_Coverage_Clean"] <= 90),

    df["Inventory_Coverage_Clean"] > 90
]

choices = [
    "Critical",
    "Healthy",
    "High Inventory",
    "Overstock"
]

df["Inventory_Status"] = np.select(
    conditions,
    choices,
    default="Unknown"
)

df["Inventory_Status"].value_counts()
```

### 2.6 Inventory Health Distribution

```python
round(
    df["Inventory_Status"]
    .value_counts(normalize=True)
    * 100,
    2
)
```

### Step 02 Findings

- 47.4% of inventory records were classified as Healthy, indicating that a large proportion of stock is aligned with forecast demand.

- 16.8% of inventory records were classified as Critical, suggesting potential replenishment risk and requiring closer monitoring.

- 34.8% of inventory records were classified as either High Inventory or Overstock, highlighting potential opportunities for stock reallocation and inventory optimisation.

- Inventory Coverage proved to be a more meaningful measure of stock health than Inventory Level alone because it considered both stock availability and forecast demand.

- Extreme coverage values were primarily driven by near-zero demand forecasts rather than unusually high inventory levels, indicating potential slow-moving inventory.

# Step 04 SKU Risk Analysis

### Objective

Identify which SKUs contribute most to critical inventory risk and overstock exposure, supporting product-level allocation and replenishment decisions.

### 4.1 SKU Inventory Status Summary

```python
sku_status_count = pd.crosstab(
    df["SKU_ID"],
    df["Inventory_Status"]
)

sku_status_count.head()
```

### 4.2 Top Critical SKUs

```python
critical_sku = (
    sku_status_count["Critical"]
    .sort_values(ascending=False)
)

critical_sku.head(10)
```

### 4.3 Critical SKU Investigation

```python
top_critical_skus = [
    "SKU_50",
    "SKU_44",
    "SKU_3",
    "SKU_30",
    "SKU_45"
]

df[
    df["SKU_ID"].isin(top_critical_skus)
].groupby("SKU_ID")[[
    "Units_Sold",
    "Demand_Forecast",
    "Inventory_Level",
    "Supplier_Lead_Time_Days"
]].median().round(2)
```

### 4.3 Critical SKU Root Cause Analysis

### Findings

The highest-risk SKUs did not exhibit unusually high demand forecasts or exceptionally long supplier lead times.

Median demand forecasts were consistent with the overall dataset (~20 units), and lead times ranged from 4 to 10 days, which was within the normal operating range.

The primary difference was inventory level. High-risk SKUs maintained lower median inventory positions (392–432 units) compared with the overall inventory median (461 units), resulting in lower inventory coverage and a higher likelihood of entering critical inventory status.

### 4.4 Critical Inventory Rate by SKU

```python
critical_rate = (
    sku_status_count["Critical"]
    / sku_status_count.sum(axis=1)
    * 100
).sort_values(ascending=False)

critical_rate.head(10)
```

### 4.5 Top Overstock SKU

```python
overstock_rate = (
    sku_status_count["Overstock"]
    / sku_status_count.sum(axis=1)
    * 100
).sort_values(ascending=False)

overstock_rate.head(10)
```

### 4.6 SKU Profile Comparison

```python
critical_rate.sort_values().head()
```

```python
critical_rate.median()
```

### 4.7 High-Risk vs Low-Risk SKU Comparison

```python
sku_compare = df[
    df["SKU_ID"].isin(["SKU_50", "SKU_16"])
]

sku_compare.groupby("SKU_ID")[[
    "Units_Sold",
    "Demand_Forecast",
    "Inventory_Level",
    "Supplier_Lead_Time_Days",
    "Inventory_Coverage_Clean"
]].median().round(2)
```

### 4.7 Findings

A comparison between SKU_50 (highest critical risk) and SKU_16 (lowest critical risk) showed that demand and sales levels were broadly similar.

The key difference was inventory position. SKU_50 maintained a median inventory level of 392 units compared with 512 units for SKU_16, resulting in lower inventory coverage (20.45 vs 25.96 days).

SKU_50 also had a longer supplier lead time (10 vs 6 days), which may further increase replenishment risk if inventory levels are not adjusted appropriately.

These findings suggest that inventory allocation levels, rather than demand patterns, are the primary driver of critical inventory risk.

### 4.8 SKU Risk Segmentation

```python
sku_risk_segment = pd.DataFrame({
    "Critical_Rate": critical_rate
})

conditions = [
    sku_risk_segment["Critical_Rate"] > 20,
    sku_risk_segment["Critical_Rate"] < 14
]

choices = [
    "High Risk",
    "Low Risk"
]

sku_risk_segment["SKU_Risk_Segment"] = np.select(
    conditions,
    choices,
    default="Normal Risk"
)

sku_risk_segment.head()
```

```python
sku_risk_segment["SKU_Risk_Segment"].value_counts()
```

### Step 04 Findings

- Inventory risk was concentrated in a small subset of products.
- Five SKUs were classified as High Risk, representing 10% of the product portfolio.
- SKU_50 recorded the highest critical inventory rate (26.1%), approximately 55% higher than the portfolio average.
- Root cause analysis indicated that lower inventory positions, rather than unusually high demand, were the primary driver of critical inventory risk.
- Overstock risk was more evenly distributed across products, suggesting that stock shortage risk was a more significant challenge than excess inventory.

Objective

Investigate whether changes in inventory coverage were driven by replenishment activity.

#### 4.9.5 Monthly Demand Forecast Trend

Objective

Investigate whether changes in inventory coverage were influenced by changes in forecast demand.

```python
monthly_forecast = (
    df.groupby("Month")["Demand_Forecast"]
    .median()
)

monthly_forecast
```

### 4.9 Trend Analysis Findings

Monthly analysis revealed a strong relationship between demand forecast, inventory coverage, and critical inventory risk.

Critical inventory events peaked in March and April, when median forecast demand reached approximately 30 units and inventory coverage fell to around 15 days.

Although replenishment activity was also highest during this period, elevated demand continued to place pressure on inventory availability.

As forecast demand declined during the second half of the year, inventory coverage increased substantially and critical inventory events fell sharply.

These findings suggest that forecast demand was the primary driver of inventory risk, highlighting the importance of proactive allocation planning and demand forecasting.

# Step 06 Power BI Data Export

Export cleaned and enriched analysis tables for Power BI dashboard development on Windows.

```python
import os

output_folder = "../powerbi_data"
os.makedirs(output_folder, exist_ok=True)
```

```python
df_powerbi = df.merge(
    sku_risk_segment[["Critical_Rate", "SKU_Risk_Segment"]],
    left_on="SKU_ID",
    right_index=True,
    how="left"
)

df_powerbi.to_csv(
    "../powerbi_data/fact_inventory_enriched.csv",
    index=False
)
```

```python
sku_risk_segment.to_csv(
    "../powerbi_data/sku_risk_summary.csv"
)
```

```python
warehouse_summary = (
    df.groupby("Warehouse_ID")
    .agg(
        Median_Coverage=("Inventory_Coverage_Clean", "median"),
        Critical_Rate=("Inventory_Status", lambda x: (x == "Critical").mean() * 100),
        Overstock_Rate=("Inventory_Status", lambda x: (x == "Overstock").mean() * 100),
        Total_Units_Sold=("Units_Sold", "sum"),
        Median_Forecast=("Demand_Forecast", "median")
    )
    .round(2)
    .reset_index()
)

warehouse_summary.to_csv(
    "../powerbi_data/warehouse_summary.csv",
    index=False
)
```

```python
monthly_trend_summary = (
    df.groupby("Month")
    .agg(
        Critical_Count=("Inventory_Status", lambda x: (x == "Critical").sum()),
        Median_Coverage=("Inventory_Coverage_Clean", "median"),
        Median_Forecast=("Demand_Forecast", "median"),
        Total_Order_Quantity=("Order_Quantity", "sum"),
        Total_Units_Sold=("Units_Sold", "sum")
    )
    .round(2)
    .reset_index()
)

monthly_trend_summary.to_csv(
    "../powerbi_data/monthly_trend_summary.csv",
    index=False
)
```

```python
action_summary_table = (
    df["Allocation_Action"]
    .value_counts(normalize=True)
    .mul(100)
    .round(2)
    .reset_index()
)

action_summary_table.columns = ["Allocation_Action", "Percentage"]

action_summary_table.to_csv(
    "../powerbi_data/action_summary.csv",
    index=False
)
```
